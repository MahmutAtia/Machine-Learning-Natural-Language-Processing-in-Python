```python
from nltk import pos_tag
```

```python
pos_tag("the girl is beautful".split())
```

```python
# function to map pos
def pos(pos_tag):
  if pos_tag.startswith("N"):
    return "n"
  elif pos_tag.startswith("V"):
    return "v"
  elif pos_tag.startswith("J"):
    return "r"
  elif pos_tag.startswith("R"):
    return "r"
  else: return "n"
```

```python
for word, tag in pos_tag("the girl is beautful, she ran to her dog after school".split()):
 print("the word :{} and the tag: {}".format(word, tag))
 print( wnl.lemmatize(word, pos=pos(tag)))
```
