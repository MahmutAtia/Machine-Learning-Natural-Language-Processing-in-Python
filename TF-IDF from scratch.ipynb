```python
# loading libraries
import nltk
from nltk import word_tokenize
nltk.download('punkt')
  
```

```python
import numpy as np
import pandas as pd
df = pd.read_csv("bbc_text_cls.csv")
```

```python
df.head()
```

```python
docs= ["i like cats i like dogs i like rats", "i like you", "you and me"]
```

```python

# creatig word to index maping
indx = 0
word2index = {}
tokenized_docs =[]
#docs = df["text"]
for text in docs:
  doc_as_int=[]
  tokens = word_tokenize(text.lower())
  for word in tokens:
    if word not in word2index:
      word2index[word]= indx
      indx+=1
      #save for later
    doc_as_int.append(word2index[word])
  tokenized_docs.append(doc_as_int)

```

```python
# reverse mapping
val_list = list(word2index.keys())
```

```python
text = ""
for i in tokenized_docs[10]:
  text += " "+val_list[i]
```

```python
# we need to create a matrixs
N=len(df["text"])
V= len(word2index)
X = np.zeros((N,V))
```

```python
for i, doc_as_int in enumerate(tokenized_docs):
  for j in doc_as_int:
    X[i,j]+=1
```
