```python
from gensim.models import KeyedVectors
```

```python
word_vectors = KeyedVectors.load_word2vec_format("/content/GoogleNews-vectors-negative300.bin", binary = True  )
```

```python
r =word_vectors.most_similar(positive=["japan"], negative=["france", "king"])
```

```python
def find_analogies(w1,w2,w4):
  words = word_vectors.most_similar(positive= [w1,w2], negative=[w4])
  print("{} - {} = {} - {}".format(w1,w2,words[0][0],w4))
```

```python
find_analogies("boy", "father", "mother")
```

```python
def nearest_word(w1):
  words = word_vectors.most_similar(positive= [w1])
  for word, score in words:
    print(word)
```

```python
nearest_word("be")
```
