## Building a function to recommend movies

```python
def recommend(vectorizer, film_docs, all_docs , film_title, films_dataframe):
  df = films_dataframe
  ind = df.loc[df["title"]== film_title].index
  for i in ind:
    film_doc = film_docs[i]
  sim_array=cosine_similarity(vectorizer.transform([film_doc]), all_docs ).flatten()
  sim_df =pd.DataFrame(sim_array)
  sim_df_sorted=sim_df.sort_values(by=0, ascending= False)
  most_five = sim_df_sorted[1:6].index
  print(df["title"][most_five])
```

```python
recommend(tfidf, film_docs, all_docs, "Pirates of the Caribbean: Dead Man's Chest", df)
```
